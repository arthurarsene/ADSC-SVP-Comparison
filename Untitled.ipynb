```python
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import cartopy.crs as ccrs
```

```python
ds = pd.read_csv('data/interpolated_gld.20221202_104507')

bathy = xr.open_dataset('GEBCO_02_Dec_2022_8bb5cd71a07e/gebco_2022_n62.0_s59.0_w-49.0_e-40.0.nc')
```

```python
idd = np.array([ds.iloc[i].values[0].split()[0] for i in range(len(ds))], dtype=int)
time = np.array([ds.iloc[i].values[0].split()[1] for i in range(len(ds))], dtype=np.datetime64)
speed = np.array([ds.iloc[i].values[0].split()[8] for i in range(len(ds))], dtype=float)
lon = np.array([ds.iloc[i].values[0].split()[4] for i in range(len(ds))], dtype=float)
lat = np.array([ds.iloc[i].values[0].split()[3] for i in range(len(ds))], dtype=float)
```

```python
plt.hist(time)
```

```python
len(np.unique(idd))
```

```python
start_time = np.array([time[idd == np.unique(idd)[i]][0] for i in range(len(np.unique(idd)))])
start_lon = np.array([lon[idd == np.unique(idd)[i]][0] for i in range(len(np.unique(idd)))])
start_lat = np.array([lat[idd == np.unique(idd)[i]][0] for i in range(len(np.unique(idd)))])
```

```python
plt.scatter(start_lon[start_time>np.datetime64('2021-08')], start_lat[start_time>np.datetime64('2021-03')])
```

```python
len(start_lon[(start_time>np.datetime64('2021-08'))&(start_time<np.datetime64('2021-09'))])
```

```python
bathy.elevation.plot(vmin=-1000, vmax=0, cmap='binary')
plt.scatter(start_lon, start_lat, s=1, c='m')
plt.xlim(-42.5, -40.5)
plt.ylim(61.7, 62)
plt.axvline(-40.9, c='lime')
```

```python
plt.figure(figsize=(16,8))
ax = plt.axes(projection=ccrs.Orthographic(-45, 60))
bathy.elevation.plot(ax=ax, transform=ccrs.PlateCarree())
ax.scatter(start_lon, start_lat, s=1, c='m', transform=ccrs.PlateCarree())

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')

ax.fill_between([-42.5, -40.9], 61.7, 62, transform=ccrs.PlateCarree(), alpha=0.3, color='lime')
```

```python
len(np.unique(idd)[(start_lon>-42.5)&(start_lon<-40.9)&(start_lat>61.7)])
```

```python
good_ids = np.unique(idd)[(start_lon>-42.5)&(start_lon<-40.9)&(start_lat>61.7)]
```

```python
plt.figure(figsize=(16,8))
ax = plt.axes(projection=ccrs.Orthographic(-45, 60))
bathy.elevation.plot(ax=ax, transform=ccrs.PlateCarree())
ax.scatter(start_lon, start_lat, s=1, c='m', transform=ccrs.PlateCarree())

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')

ax.fill_between([-42.5, -40.9], 61.7, 62, transform=ccrs.PlateCarree(), alpha=0.3, color='lime')

for i in range(len(good_ids)):
    ax.plot(lon[idd==good_ids[i]], lat[idd==good_ids[i]], transform=ccrs.PlateCarree())
```
